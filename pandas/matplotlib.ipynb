# 标记注释

## 图形标记

* 在数据可视化的过程中，图片中的文字经常被用来注释图中的一些特征。使用annotate()方法可以很方便地添加此类注释。在使用annotate时，要考虑两个点的坐标：被注释的地方xy(x, y)和插入文本的地方xytext(x, y)。
* 详细参数见

    https://matplotlib.org/api/_as_gen/matplotlib.pyplot.annotate.html?highlight=annotate#matplotlib.pyplot.annotate

```python
plt.plot(ridership_df.index, ridership_df['R003'], 'r+:', label='R003')  # 默认参数设定颜色
plt.legend(loc=0)

# fontsize: 注释字体大小

# arrowprops描述箭头的属性
    # facecolor:颜色，
    # frac:箭头头部所占据的比例，
    # headwidth:箭头宽度
    # width:箭身宽度，
    # arrowstyle : 箭头样式 
    #              '->'指向标注点 
    #           '<-'指向标注内容 还有很多'-' 

plt.annotate('local max', xy=(ridership_df.index[3], ridership_df['R003'][3]), xytext=(ridership_df.index[3], ridership_df['R003'][3]-500),
             fontsize=16, arrowprops=dict(facecolor='g', arrowstyle='->')
            )
```

## 文字标记

    在图中显示文字说明

```python
plt.plot(ridership_df.index, ridership_df['R003'], 'r+:', label='R003')  # 默认参数设定颜色
plt.legend(loc=0)

# text 参数说明（依照前后顺序，前三个参数顺序不变，后面按需要选择）：
#    文字的起始坐标位置
#    文字内容
#    family: 字体（主要是英文字体）
#    size: 字体大小
#    color: 颜色
#    style：italic(斜体) ...
#    weight: 粗细程度（0~100）
#    bbox: 文字加框
#         facecolor: 框颜色
#         alpha: 透明度(0~1)

plt.text(ridership_df.index[3], 800, u'一段文字注释')
plt.text(ridership_df.index[3], 1000, u'又来一个加框的', bbox=dict(facecolor='g', alpha=0.5))
```

## 双坐标轴

```python
# 示例数据
df1 = pd.DataFrame(
    data=[[2, 3877],
          [4, 4088],
          [16, 3392],
          [33, 4802]],
    index=['05-01-11', '05-02-11', '05-03-11', '05-04-11'],
    columns=['R003', 'R004']
)
df1.index = pd.to_datetime(df1.index)
df1
```

```python
import matplotlib as mpl  
mpl.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签  
mpl.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
fig = plt.figure() # 创建图对象

# 增加一个子图到图对象中，图中共有2*2个子图位置，当前子图添加到第一个位置
ax1 = fig.add_subplot(111) 

# 在子图上绘图
ax1.plot(df1.index, df1['R003'], 'g.-',label='R003')
ax1.set_ylabel('R003')

# 增加一个坐标轴在当前位置，twinx函数是复制一个x坐标轴
# 因为当前图需要2个y轴和一个共同的x轴
ax2 = ax1.twinx()
ax2.plot(df1.index, df1['R004'], 'r.-', label='R004')
ax2.set_ylabel('R004')

ax1.set_xlabel(u'时间')
# 坐标格式自适应（日期太长不会挤在一起）
fig.autofmt_xdate()
```

```python
# 也可以绘制双x轴，与上面双y轴类似

fig = plt.figure() # 创建图对象

# 增加一个子图到图对象中，图中共有2*2个子图位置，当前子图添加到第一个位置
ax1 = fig.add_subplot(111) 

# 在子图上绘图
ax1.plot(df1['R003'], df1.index,'g.-',label='R003')
ax1.set_xlabel('R003')

# 增加一个坐标轴在当前位置，twiny函数是复制一个y坐标轴
# 因为当前图需要2个x轴和一个共同的y轴
ax2 = ax1.twiny()
ax2.plot(df1['R004'], df1.index, 'r.-', label='R004')
ax2.set_xlabel('R004')

ax1.set_ylabel('Time')
# 坐标格式自适应（日期太长不会挤在一起）
fig.autofmt_xdate()
```
